```python
# ШАГ 2: Импорт библиотек
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from flaml import AutoML
import warnings
warnings.filterwarnings('ignore')

print("Библиотеки импортированы!")
```

```python
# ШАГ 3: Загрузка данных
print("ЗАГРУЗКА ДАННЫХ")

from google.colab import files

print("загрузите файл 'disparenzarization_data_2026.csv'")
uploaded = files.upload()

# Читаем файл
import io
for filename in uploaded.keys():
    df = pd.read_csv(io.BytesIO(uploaded[filename]))
    print(f" Файл загружен! Размер: {df.shape[0]} строк, {df.shape[1]} столбцов")
```

```python
# ШАГ 4: Быстрый анализ данных

print("АНАЛИЗ ДАННЫХ")

print("Первые 5 строк данных:")
print(df.head())

print("\nИнформация о данных:")
print(f"Всего признаков: {df.shape[1] - 1}")  # минус целевая переменная
print(f"Целевая переменная 'Доклинический_риск':")
print(df['Доклинический_риск'].value_counts())
print(f"\nДоля пациентов с риском: {df['Доклинический_риск'].mean()*100:.2f}%")

# Проверяем пропуски
print(f"\nПропуски в данных: {df.isnull().sum().sum()}")
if df.isnull().sum().sum() > 0:
    print("Столбцы с пропусками:")
    print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# ШАГ 5: Подготовка данных

print(" ПОДГОТОВКА ДАННЫХ")


# Отделяем признаки (X) от целевой переменной (y)
X = df.drop('Доклинический_риск', axis=1)
y = df['Доклинический_риск']

print(f"Признаки (X): {X.shape}")
print(f"Целевая переменная (y): {y.shape}")

# Разделяем на обучающую (80%) и тестовую (20%) выборки
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 20% данных для теста
    random_state=42,    # чтобы результат можно было воспроизвести
    stratify=y          # сохраняем пропорцию классов
)

print(f"\nОбучающая выборка: {X_train.shape[0]} пациентов")
print(f"Тестовая выборка: {X_test.shape[0]} пациентов")
print(f"Доля риска в обучении: {y_train.mean()*100:.2f}%")
print(f"Доля риска в тесте: {y_test.mean()*100:.2f}%")
```

```python
# ШАГ 6: Предобработка (заполнение пропусков и нормализация)

print("ПРЕДОБРАБОТКА ДАННЫХ")


# Заполняем пропуски медианой
imputer = SimpleImputer(strategy='median')
X_train_clean = imputer.fit_transform(X_train)
X_test_clean = imputer.transform(X_test)

# Нормализуем данные
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_clean)
X_test_scaled = scaler.transform(X_test_clean)

# Преобразуем обратно в DataFrame для удобства
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X.columns)

print(" Пропуски заполнены, данные нормализованы")
```

```python
# ШАГ 7: ОБУЧЕНИЕ МОДЕЛЕЙ

print("ОБУЧЕНИЕ AutoML МОДЕЛЕЙ")


# Словарь для хранения результатов
results = {}


# МОДЕЛЬ 1: FLAML (быстрая)

print("\n Модель 1: FLAML (оптимизированная)")


automl_flaml = AutoML()
automl_flaml.fit(
    X_train_scaled, y_train,
    task='classification',
    time_budget=120,  # 2 минуты
    metric='roc_auc',
    verbose=0
)

# Предсказания
y_pred_flaml = automl_flaml.predict(X_test_scaled)
y_proba_flaml = automl_flaml.predict_proba(X_test_scaled)[:, 1]

# Метрики
results['FLAML'] = {
    'accuracy': accuracy_score(y_test, y_pred_flaml),
    'precision': precision_score(y_test, y_pred_flaml, zero_division=0),
    'recall': recall_score(y_test, y_pred_flaml),
    'f1': f1_score(y_test, y_pred_flaml),
    'roc_auc': roc_auc_score(y_test, y_proba_flaml)
}

print(f" FLAML обучена!")
print(f"   ROC-AUC: {results['FLAML']['roc_auc']:.3f}")
print(f"   F1-Score: {results['FLAML']['f1']:.3f}")
```

```python
# МОДЕЛЬ 2: Random Forest (через FLAML)

print("\n Модель 2: Random Forest (только случайный лес)")

automl_rf = AutoML()
automl_rf.fit(
    X_train_scaled, y_train,
    task='classification',
    time_budget=90,
    metric='roc_auc',
    estimator_list=['rf'],  # только случайный лес
    verbose=0
)

# Предсказания
y_pred_rf = automl_rf.predict(X_test_scaled)
y_proba_rf = automl_rf.predict_proba(X_test_scaled)[:, 1]

# Метрики
results['Random Forest'] = {
    'accuracy': accuracy_score(y_test, y_pred_rf),
    'precision': precision_score(y_test, y_pred_rf, zero_division=0),
    'recall': recall_score(y_test, y_pred_rf),
    'f1': f1_score(y_test, y_pred_rf),
    'roc_auc': roc_auc_score(y_test, y_proba_rf)
}

print(f" Random Forest обучена!")
print(f"   ROC-AUC: {results['Random Forest']['roc_auc']:.3f}")
print(f"   F1-Score: {results['Random Forest']['f1']:.3f}")
```

```python
# МОДЕЛЬ 3: LightGBM (через FLAML)

print("\n Модель 3: LightGBM")

automl_lgbm = AutoML()
automl_lgbm.fit(
    X_train_scaled, y_train,
    task='classification',
    time_budget=90,
    metric='roc_auc',
    estimator_list=['lgbm'],  # только LightGBM
    verbose=0
)

# Предсказания
y_pred_lgbm = automl_lgbm.predict(X_test_scaled)
y_proba_lgbm = automl_lgbm.predict_proba(X_test_scaled)[:, 1]

# Метрики
results['LightGBM'] = {
    'accuracy': accuracy_score(y_test, y_pred_lgbm),
    'precision': precision_score(y_test, y_pred_lgbm, zero_division=0),
    'recall': recall_score(y_test, y_pred_lgbm),
    'f1': f1_score(y_test, y_pred_lgbm),
    'roc_auc': roc_auc_score(y_test, y_proba_lgbm)
}

print(f" LightGBM обучена!")
print(f"   ROC-AUC: {results['LightGBM']['roc_auc']:.3f}")
print(f"   F1-Score: {results['LightGBM']['f1']:.3f}")
```

```python
# ШАГ 8: СРАВНЕНИЕ РЕЗУЛЬТАТОВ

print("СРАВНЕНИЕ МОДЕЛЕЙ")


# Создаем таблицу результатов
results_df = pd.DataFrame(results).T
results_df = results_df.round(4)
results_df = results_df.sort_values('roc_auc', ascending=False)

print("\n СРАВНИТЕЛЬНАЯ ТАБЛИЦА МЕТРИК:")

print(results_df.to_string())


# Находим лучшую модель
best_model_name = results_df.index[0]
print(f"\n ЛУЧШАЯ МОДЕЛЬ: {best_model_name}")
print(f"   ROC-AUC: {results_df.loc[best_model_name, 'roc_auc']:.4f}")
print(f"   Accuracy: {results_df.loc[best_model_name, 'accuracy']:.4f}")
print(f"   F1-Score: {results_df.loc[best_model_name, 'f1']:.4f}")



```
